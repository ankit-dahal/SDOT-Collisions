## Chi-square test and Kolmogorov-Smirnov test



---

`Chi-square test` for `SPEEDING` and `SEVERITYDESC`, `INATTENTIONIND` and `SEVERITYDESC`, `UNDERINFL` and `SEVERITYDESC`, `WEATHER` and `SEVERITYDESC`, `COLLISIONTYPE` and `SEVERYITYDESC` show strong statistical relationship.

---



````python
from scipy.stats import chi2_contingency, kstest

contingency_table = pd.crosstab(new_df['SPEEDING'], new_df['SEVERITYDESC'])

chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-square test statistic:", chi2)
print("p-value:", p)
````

````python
contingency_table = pd.crosstab(new_df['INATTENTIONIND'], new_df['SEVERITYDESC'])

chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-square test statistic:", chi2)
print("p-value:", p)
````

````python
contingency_table = pd.crosstab(new_df['UNDERINFL'], new_df['SEVERITYDESC'])

chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-square test statistic:", chi2)
print("p-value:", p)
````

````python
contingency_table = pd.crosstab(new_df['WEATHER'], new_df['SEVERITYDESC'])

chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-square test statistic:", chi2)
print("p-value:", p)
````

````python
contingency_table = pd.crosstab(new_df['COLLISIONTYPE'], new_df['SEVERITYDESC'])

chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-square test statistic:", chi2)
print("p-value:", p)
````



---

Using `Kolmogorov-Smirnov test` for `VEHCOUNT`, `PERSONCOUNT` and `INJURIES`, none of these data are shown to be normal.

---



````python
ks_speeding_stat, ks_speeding_p_value = kstest(new_df['VEHCOUNT'], 'norm', args=(new_df['VEHCOUNT'].mean(), new_df['VEHCOUNT'].std()))

print("\nKolmogorov-Smirnov Test for VEHCOUNT:")
print("Test Statistic:", ks_speeding_stat)
print("P-value:", ks_speeding_p_value)
````

````python
ks_speeding_stat, ks_speeding_p_value = kstest(new_df['PERSONCOUNT'], 'norm', args=(new_df['PERSONCOUNT'].mean(), new_df['PERSONCOUNT'].std()))

print("\nKolmogorov-Smirnov Test for PERSONCOUNT:")
print("Test Statistic:", ks_speeding_stat)
print("P-value:", ks_speeding_p_value)
````

````python
ks_speeding_stat, ks_speeding_p_value = kstest(new_df['INJURIES'], 'norm', args=(new_df['INJURIES'].mean(), new_df['INJURIES'].std()))

print("\nKolmogorov-Smirnov Test for INJURIES:")
print("Test Statistic:", ks_speeding_stat)
print("P-value:", ks_speeding_p_value)
````

## Conclusion

Conclusions drawn from each univariate, bivariate or multivariate data analysis are mentioned along with the plots. Some of them are:
*   Majority of the collisions takes place in `BLOCKS`
*   Collision related to `PARKED CAR`, `ANGLES` and `REAR ENDED` are significantly higher than other types of collisions.
*   Most of the incidents involve `NO` pedestrians.
*   Most of the incidents involve no `INJURIES`.
*   `Injury Collision` is frequent during `SPEEDING` than when not speeding.
*   When `Pedestrian` and `Cycles` are involved, there is high change of `Serious Injury`.
*   Probability of Injury is higher under influence than not under influence
*   When only single vehicle is involved, there is a greater chance of a serious injury than when multiple vehicles are involved.
*   Most of the collisions occur during `Daylight` or during `Dark - Street Lights Off`
*   `Speeding`, `Under Influence`, `Inattention`, `Weather` and `Collision Type` are directly involved with `Severity`

To predict collision severity, the feature I could identify were `INATTENTIONIND`, `UNDERINFL`, `SPEEDING`, `WEATHER`, `COLLISIONTYPE`.
i.e. If the person involved is `not paying attention` or `under influence of any drug or alcohol` or `speeding` or `driving in bad weather`, then they're likely to be involved in a collision.
